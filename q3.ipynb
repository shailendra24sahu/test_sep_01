```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('stock_prices.csv')
df
```

```python
# Parse the date into datetime format
df['date'] = pd.to_datetime(df['date'])

# Set the 'date' column as the index
df.set_index('date', inplace=True)

# Calculate the rolling 30-day moving average
rolling_avg = df['closing_price'].rolling(window=30).mean()
```

```python
df
```

```python
# Create a plot for the closing price and rolling average
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['closing_price'], label='Closing Price')
plt.plot(df.index, rolling_avg, label='30-Day Moving Average', linestyle='--', color='orange')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price Analysis with 30-Day Moving Average')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
